```python
# genrating surname
df.insert(3,"surname",df["Name"].str.strip().str.split(",").str.get(0))
```

```python
# genrating Title eg. Mr ,Mrs
df.insert(4,"title",df["Name"].str.strip().str.split(",").str.get(1).str.split(".").str.get(0))
```

```python
# genrating name 
df.insert(5,"name",df["Name"].str.strip().str.split(",").str.get(1).str.split(".").str.get(1).str.split("(").str.get(0))
```

```python
# Name Column is use less
df.drop( ["Name"] , axis = 1 , inplace = True)
```
